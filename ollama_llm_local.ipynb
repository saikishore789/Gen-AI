```python
import ollama
response = ollama.chat(
    model="llama3.2:1b",
    messages=[
        {"role": "system", "content": "you are a assistnant"},
        {"role": "user", "content": "what is ip address"}
]
)

print(response["message"]["content"])
```

```python
import ollama

stream = ollama.chat(
    model="llama3.2:1b",
    messages=[
        {"role": "system", "content": "You are a helpful assistant"},
        {"role": "user", "content": "What is Artificial Intelligence?"}
    ],
    stream=True
)

for chunk in stream:
    print(chunk["message"]["content"], end="", flush=True)
```
