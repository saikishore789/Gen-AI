```python
from google import genai

client = genai.Client()

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explain how kubernetes works in a few words"
)

print(response.text)
```

```python
# multi-turn conversations (chat)

from google import genai

client = genai.Client()

chat = client.chats.create(model="gemini-2.5-flash")

response = chat.send_message("I need 1 application to be deployed in kubernetes")
print(response.text)

print("-"*50)

response = chat.send_message("How many nodes do i need to create ")
print(response.text)
```
